```python
from pypdf import PdfReader

def extract_text(pdf_path):
  reader = PdfReader(pdf_path)
  text = " "
  for page in reader.pages:
    text+= page.extract_text() + "/n "
  return text

text = extract_text('/content/Finance_Bill_2025.pdf')
```

```python
text
```

```python
# chunk - split
def chunk_text(text, chunk_size = 800):
  words = text.split()
  chunks = []
  for i in range(0,len(words), chunk_size):
    chunks.append(" ".join(words[i:i+chunk_size]))
  return chunks
```

```python
chunks = chunk_text(text)
```

```python
chunks
```

```python
words = text.split()
words
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

model = SentenceTransformer('BAAI/bge-small-en-v1.5')
```

```python
finance_embeddings = model.encode(chunks).tolist()
```

```python
len(finance_embeddings)
```

```python
import chromadb
```

```python
client = chromadb.PersistentClient(path="./chroma_store")
```

```python
collection =client.create_collection(
    name = 'finance_collections',
    metadata = {'description':'the_finance_collection'}
)

print('collection_ created:',collection.name)
```

```python
ids = [f'doc_{i}' for i in range(len(chunks))]
```

```python
ids
```

```python
len(chunks) == len(ids)
```

```python
collection.add(
    documents = chunks,
    embeddings = finance_embeddings,
    ids = ids
)
```

```python
collection.count()
```

```python
query = "What does the finance bill say about taxes"
```

```python
query_embedding = model.encode(query).tolist()
```

```python
query_embedding
```

```python
results = collection.query(
    query_embeddings = query_embedding,
    n_results = 3
)
```
